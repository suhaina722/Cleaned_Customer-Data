```python
import pandas as pd

df = pd.read_csv("Customer Behaviour Analysis.csv")
print(df.head())
```

```python
# Remove spaces from column names
df.columns = df.columns.str.strip()
```

```python
#check missing values
df.isnull().sum()
```

```python
print(df.columns)
```

```python
#Fill rating with median
df['review_rating']=df['review_rating'].fillna(df['review_rating'].median())
```

```python
print(list(df.columns))
```

```python
#Fill categorial columns with 'Unknown'
categorical_cols = df.select_dtypes(include = ['object']).columns
```

```python
#standardize column names(snake_case)
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
#Remove redundant columns
df.drop(columns=['promo_code_used'], inplace=True,errors = 'ignore')
```

```python
#Age Group
def age_group(age):
  if age < 25:
    return "Young"
  elif age < 45:
    return "Adult"
  else:
    return "senior"
df['age_group'] = df ['age'].apply(age_group)
```

```python
#purchase_frequency_days
def freq_to_days(freq):
  if freq == 'weekly':
    return 7
  elif freq == 'monthly':
    return 30
  elif freq == 'Annually': # Added this condition based on 'Annually' in df.head()
    return 14
  else:
    return 365



df['purchase_frequency_days'] = df['frequency_of_purchases'].apply(freq_to_days)
```

```python
#Validate Data Consistency
#check duplicates
df.duplicated().sum()
```

```python
#Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
#Check data types
df.dtypes
```

```python
#Summery Statistics
df.describe()
```

```python
#clean dataset
#save cleaned dataset
df.to_csv("cleaned_customer_data.csv",index=False)
```
